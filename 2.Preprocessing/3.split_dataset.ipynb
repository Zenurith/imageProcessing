```python
import json
import os
import random
import shutil

# ======================
# PATHS
# ======================

IMAGE_DIR = r"C:\Users\User\Downloads\archive\processed_images"
ANNOTATION_FILE = r"C:\Users\User\Downloads\archive\annotations_resized.json"

OUTPUT_DIR = r"C:\Users\User\Downloads\archive\dataset"

TRAIN_RATIO = 0.7
VAL_RATIO = 0.2
TEST_RATIO = 0.1

# ======================
# LOAD DATA
# ======================

with open(ANNOTATION_FILE) as f:
    data = json.load(f)

images = data["images"]

# ======================
# SHUFFLE IMAGES
# ======================

random.shuffle(images)

total = len(images)

train_end = int(total * TRAIN_RATIO)
val_end = int(total * (TRAIN_RATIO + VAL_RATIO))

train_images = images[:train_end]
val_images = images[train_end:val_end]
test_images = images[val_end:]

print("Train:", len(train_images))
print("Val:", len(val_images))
print("Test:", len(test_images))

# ======================
# CREATE FOLDERS
# ======================

for split in ["train", "val", "test"]:
    os.makedirs(os.path.join(OUTPUT_DIR, split), exist_ok=True)

# ======================
# COPY IMAGES
# ======================

def copy_images(image_list, split):

    for img in image_list:

        file_name = img["file_name"].replace("/", "_")

        src = os.path.join(IMAGE_DIR, file_name)
        dst = os.path.join(OUTPUT_DIR, split, file_name)

        if os.path.exists(src):
            shutil.copy(src, dst)

copy_images(train_images, "train")
copy_images(val_images, "val")
copy_images(test_images, "test")

print("Dataset split complete.")
```
