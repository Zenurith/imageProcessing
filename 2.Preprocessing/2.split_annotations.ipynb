```python
import json
import os

ANNOTATION_FILE = r"C:\Users\User\Downloads\archive\annotations_resized.json"
DATASET_DIR = r"C:\Users\User\Downloads\archive\dataset"

with open(ANNOTATION_FILE) as f:
    data = json.load(f)

images = data["images"]
annotations = data["annotations"]
categories = data["categories"]

# build lookup
image_lookup = {img["file_name"].replace("/", "_"): img["id"] for img in images}

def create_split(split):

    folder = os.path.join(DATASET_DIR, split)

    image_ids = []

    for file in os.listdir(folder):
        if file in image_lookup:
            image_ids.append(image_lookup[file])

    split_images = [img for img in images if img["id"] in image_ids]
    split_annotations = [ann for ann in annotations if ann["image_id"] in image_ids]

    split_data = {
        "images": split_images,
        "annotations": split_annotations,
        "categories": categories
    }

    output_path = os.path.join(DATASET_DIR, f"{split}_annotations.json")

    with open(output_path, "w") as f:
        json.dump(split_data, f)

    print(split, "annotations saved.")

create_split("train")
create_split("val")
create_split("test")
```
